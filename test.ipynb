```python
sch = [[(('M1', 'W1'), ('M2', 'W2')), (('M3', 'W3'), ('M4', 'W4'))], 
            [(('M1', 'W2'), ('M3', 'W4')), (('M4', 'W1'), ('M2', 'W3'))], 
            [(('M1', 'W3'), ('M4', 'W2')), (('M2', 'W4'), ('M3', 'W1'))], 
            [(('M1', 'W4'), ('M2', 'W1')), (('M3', 'W2'), ('M4', 'W3'))]]
```

```python
history = list()
round_num = 1
for round in sch:
    for match in round:
        ((A, B), (C, D)) = match
        history.append(round_num, A, B, C, D)
        history.append(round_num, B, A, C, D)
        history.append(round_num, C, D, A, B)
        history.append(round_num, D, C, A, B)
    round_num += 1
```

```python
n = 4
rounds=4
men = [f"M{i+1}" for i in range(n)]
women = [f"W{i+1}" for i in range(n)]

schedule = []

for r in range(rounds):
    n_len = len(men)

    # Rotate women to create unique partners
    rotated_women = women[r%n_len:] + women[:r%n_len]

    teams = list(zip(men, rotated_women))
    schedule.append(teams)
for i in schedule:
    print(i)
```
